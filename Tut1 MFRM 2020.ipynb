```python
# Exercise 1
somelist = [i for i in range(1,101)]
mean = sum(somelist)*(1/len(somelist))
print('mean of somelist is', mean)

somelist_squared = [x ** 2 for x in somelist]
variance = sum(somelist_squared)*(1/len(somelist)) - mean ** 2
print('variance of somelist is', variance)
```

```python
# Exercise 2
for i in range(1,101):
    if i % 3 == 0 and i % 5 == 0:
        print('fizzbuzz')
    elif i % 3 == 0:
        print('fizz')
    elif i % 5 == 0:
        print('buzz')
    else:
        print(i)
```

```python
# Exercise 3
def f(n):
    if n % 2 == 0:
        return n // 2
    else:
        return 3*n+1

for i in range(1, 101):
    n = i
    n = f(n) 
    reached_1 = (n == 1)
    while not reached_1:
        n = f(n)
        reached_1 = (n == 1)
    print('checked f for n=', i)
```
